1. 登录并进入子账户
点击你界面上的 “转到您的试用账户” (Go To Your Trial Account)。

点击进入你的 试用子账户 (Trial Subaccount)。








2. 订阅 Integration Suite
在左侧菜单栏中，找到 “服务” (Services) -> “实例和订阅” (Instances and Subscriptions)。





点击 “创建” (Create)。

在服务列表中搜索并选择 “Integration Suite”。

选择对应的计划（通常为 trial），点击 “创建”。订阅过程可能需要几分钟。






3. 启用集成功能
订阅完成后，点击实例列表中的 Integration Suite 链接以打开该应用程序。

进入后，你需要 “添加功能” (Add Capabilities)。

勾选并启用 “Cloud Integration”（这就是你要的 CPI），也可以同时勾选 API Management 或其他你感兴趣的模块。

点击保存并等待激活（激活过程通常需要 5-10 分钟）。

4. 分配角色 (非常重要)
启用功能后，如果你直接访问仍然会遇到权限错误。你需要为你自己分配管理权限：

回到 BTP Cockpit 的子账户视图。

点击左侧的 “安全性” (Security) -> “用户” (Users)。

选择你的用户，点击 “分配角色集” (Assign Role Collection)。

搜索并添加如 Integration_Provisioner、PI_Administrator 等相关角色集，以便你可以进行配置和部署。
